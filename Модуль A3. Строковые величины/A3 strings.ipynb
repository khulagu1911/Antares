```python

#Формируем паттерн для выгрузки оценок
import re
pattern = re.compile('\d+')

exams = []

f = open('StudentPerformance.csv')
#Считываем файл исключая заголовки в список 
for line in f:
    info = line.split(',')
#    print(info[0])
    if info[0] == '"gender':
        continue
#Добавляем строки с форматированием оценок в числа
    else:
        new_line = []
        for item in info:
            if pattern.search(item) != None:
                new_line.append(int(pattern.search(item)[0]))
            else:
                new_line.append(item[2:-2])
        exams.append(new_line)
        
print(exams)
```
