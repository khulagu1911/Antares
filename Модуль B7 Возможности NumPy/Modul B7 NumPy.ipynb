```python
import numpy as np
import pandas as pd

first_line = [x*y for x in range(2, 100, 6) for y in range (7, 1, -2)]
second_line = [x ** 0.5 for x in range(1000, 1101, 2)]
third_line = [x**2 for x in range(51)]
big_secret = np.array([first_line, second_line, third_line, second_line, first_line])

x_row = big_secret.shape[0]
x_col = big_secret.shape[1]
for i in range(1, x_row, 2):
    for j in range(1, x_col, 2):
        big_secret[i, j] = 1
for i in range(0, x_row, 2):
    for j in range(0, x_col, 2):
        big_secret[i, j] = -1
display(big_secret)
```
